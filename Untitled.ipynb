```python
from PIL import ImageFont,ImageDraw,Image
from barcode.writer import ImageWriter
barcode_format=barcode.get_barcode_class('code128')
barCode='V000547'
writer=ImageWriter()

generated=barcode_format(barCode,writer)
generated.save(f"myapp/static/template/bCode/{barCode}",{"module_width":0.35, "module_height":10, "quiet_zone": 3})
pil_image=Image.open(f"myapp/static/template/bCode/{barCode}.png")
crop_image=pil_image.crop((33,9,410,132))

crop_image=crop_image.resize((897,259))
template=Image.open("myapp/static/template/balaji/balaji_template.png")
template.paste(crop_image,(42,1280,939,1539))
```
